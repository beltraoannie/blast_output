```python
import pandas as pd
```

```python
nome_arquivo = "utfmt"
arq = open('/home/beltraoannie/Downloads/Cdd_LE/'+nome_arquivo)
arq_string = arq.read()
arq.close()

#Quebra o arquivo em alinhamentos individuais
alinhamentos = arq_string.split("Query=")

del alinhamentos[0]
```

```python
len(alinhamentos)
```

```python
domains = []
for alinhamento in alinhamentos:
    alinhamento_individual = alinhamento.split('\n')
    for i in alinhamento_individual:
        if 'CDD:' in i:
            domains.append(i)
            break
        if 'No hits found' in i:
            domains.append(i)
```
